```python
# dataframe employee
import pandas as pd

data = {
    "employee_id": [101, 102, 103, 104, 105],
    "name": ["Andi", "Budi", "Citra", "Dewi", "Eka"],
    "department": ["IT", "HR", "Finance", "IT", "HR"],
    "salary": [7000, 5000, 6000, 6500, 5500]
}

employees = pd.DataFrame(data)
print(employees)
```

```python
employees.loc[employees['department'] == 'IT', 'salary'] += 500
display(employees)
```

```python
# membuat kolom baru bernama "bonus" dengan salary * 0.1
employees_bonus = employees.assign(bonus=employees['salary'] * 0.1)

# menampilkan dataframe yang baru
print("DataFrame baru setelah menggunakan .assign():")
display(employees_bonus)

# menampilkan dataframe asli
print("\nDataFrame asli setelah menggunakan .assign():")
display(employees)
```

```python
ambil_data_103 = employees[employees['employee_id'] == 103]
display(ambil_data_103)
```

```python
# Buat DataFrame performance score
data_performance = {
    "employee_id": [101, 102, 103, 104, 105],
    "performance_score": [90, 75, 80, 85, 70]
}
performance_df = pd.DataFrame(data_performance)

print("DataFrame Performance Score:")
display(performance_df)
```

```python
merged = pd.merge(employees, performance_df, on="employee_id")
print(merged)
```
