**Mini project: Analisis Data Penjualan E-commerce ðŸ›’**

Alurnya:

1. Bikin DataFrame dari data dummy (misalnya: order_id, customer, product, quantity, price).

2. Hitung total harga per transaksi (quantity Ã— price).

3. Tambah kolom kategori produk (misalnya: elektronik, fashion, food).

4. Filter transaksi dengan nilai di atas Rp500.000.

5. GroupBy kategori produk â†’ hitung total penjualan per kategori.

6. (Optional) Merge dengan DataFrame lain, misalnya data customer (nama, kota), biar lebih kaya datanya.

**Output akhirnya bisa kayak laporan mini:**

a. Top 3 customer dengan belanja terbesar.

b. Penjualan per kategori.

c. Rata-rata order value.

```python
#1
import pandas as pd

data = {
    "order_id": [101, 102, 103, 104, 105],
    "customer": ["Andi", "Budi", "Citra", "Dewi", "Eka"],
    "customer_id": ["c005", "c010", "c058", "c502", "c505"],
    "product": ["minyak", "telur", "iphone", "sunscreen", "stratocaster"],
    "quantity": [7, 5, 1, 2, 1],
    "price": [70000, 10000, 6000000, 65000, 5500000]
}

data_penjualan = pd.DataFrame(data)
print(data_penjualan)
```

```python
#2
data_penjualan['total_price'] = data_penjualan['quantity'] * data_penjualan['price']
display(data_penjualan)
```

```python
#3
data_penjualan['category'] = ['makanan', 'makanan', 'elektronik', 'kesehatan', 'elektronik']

new_column_order = ['order_id', 'customer', 'customer_id', 'product', 'category', 'quantity', 'price', 'total_price']

data_penjualan = data_penjualan[new_column_order]
display(data_penjualan)
```

```python
#4
filter_transaction = data_penjualan[data_penjualan['total_price'] > 500000]
display(filter_transaction)
```

```python
#5
group_category = data_penjualan.groupby('category')['total_price'].sum()
display(group_category)
```

```python
#6
data_customer = {
    "customer_id": ["c005", "c010", "c058", "c502", "c505"],
    "customer": ["Andi", "Budi", "Citra", "Dewi", "Eka"],
    "city": ["jambi", "jakarta", "jakarta", "surabaya", "bogor"],
    "city_id": ["ct09", "ct01", "ct01", "ct10", "ct03"]
}

dt_customer = pd.DataFrame(data_customer)
print(dt_customer)
```

```python
#6
merged_dt = pd.merge(data_penjualan.drop(columns=["customer"]), dt_customer, on="customer_id", how="inner")

merged_dt = merged_dt[
    ["order_id", "customer", "customer_id", "city", "city_id",
     "product", "category", "quantity", "price", "total_price"]]

display(merged_dt)
```

# **Laporan mini project:**

## Top 3 sales

```python
#a
top_3_sales = merged_dt.sort_values(by='total_price', ascending=False).head(3)
display(top_3_sales)
```

## Top sales per category

```python
#b
top_sales_per_category = merged_dt.groupby('category')['total_price'].sum().sort_values(ascending=False)
display(top_sales_per_category)
```

## Average order value

```python
#c
average_order_value = merged_dt['total_price'].mean()
print(f"average order value: {average_order_value}")
```
